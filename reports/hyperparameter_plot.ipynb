```python
import matplotlib.pyplot as plt


models = ["RoBERTa(Large)-BiGRU", "BERT(Large)-BiGRU"]
batch_sizes = [4, 8, 16, 32, 64, 128]

roberta_values = {
    4: [0.8563, 0.8478],
    8: [0.8423, 0.8427],
    16: [0.8385, 0.8254],
    32: [0.8443, 0.8368],
    64: [0.8316, 0.8212],
    128: [0.8266, 0.7938]
}

bert_values = {
    4: [0.853, 0.8475],
    8: [0.8457, 0.8394],
    16: [0.835, 0.82],
    32: [0.8226, 0.8111],
    64: [0.8333, 0.8276],
    128: [0.8176, 0.7795]
}


plt.figure(figsize=(10, 6))

for model_values, model_name in zip([roberta_values, bert_values], models):
    plt.plot(batch_sizes, [model_values[batch][0] for batch in batch_sizes], marker='o', label=model_name + " Accuracy")
    plt.plot(batch_sizes, [model_values[batch][1] for batch in batch_sizes], marker='o', label=model_name + " F1-Score")

plt.title("Accuracy and F1-Score with varying batch size", fontsize=18)
plt.xlabel("Batch Size", fontsize=18)
plt.ylabel("Value", fontsize=18)
plt.xticks(batch_sizes, fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.legend(fontsize=14)
plt.show()
```

```python
hidden_dims = [100, 200, 256, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200]
accuracy_means = [0.8324,  0.843, 0.8530, 0.848, 0.848, 0.8425, 0.8369, 0.8233, 0.8362, 0.829, 0.8224, 0.8179, 0.7967]
f1score_means = [0.8273, 0.8384, 0.8475, 0.8436, 0.8468, 0.8319, 0.8332, 0.8192, 0.8269, 0.8221, 0.8009, 0.8062, 0.7619]


plt.figure(figsize=(10, 6))
plt.plot(hidden_dims, accuracy_means, marker='o', label='Accuracy')
plt.plot(hidden_dims, f1score_means, marker='o', label='F1-Score')


plt.title('Performance Metrics vs Hidden Dimension', fontsize=18)
plt.xlabel('Hidden Dimension', fontsize=18)
plt.ylabel('Value', fontsize=18)
plt.legend(fontsize=15)


hidden_dims_filtered = [dim for dim in hidden_dims if dim != 256]
plt.xticks(hidden_dims_filtered, fontsize=15)


plt.yticks(fontsize=15)
plt.grid(True)


plt.show()
```
